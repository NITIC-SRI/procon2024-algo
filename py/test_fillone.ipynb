```python
from board.board import Board
from board.fillone_minimal import compress_actions

start = Board(
    [
        [3, 2, 2, 3, 1, 0, 0, 3, 3, 1],
        [0, 2, 0, 2, 1, 3, 3, 0, 3, 0],
        [2, 0, 1, 2, 1, 3, 0, 1, 2, 0],
        [0, 1, 3, 1, 3, 2, 3, 3, 0, 2],
        [0, 1, 0, 1, 0, 2, 1, 1, 3, 0],
        [3, 0, 3, 3, 2, 3, 0, 2, 0, 1],
        [3, 0, 0, 1, 2, 3, 3, 2, 2, 2],
        [2, 0, 3, 0, 2, 1, 3, 1, 2, 0],
        [1, 2, 3, 3, 0, 2, 2, 2, 3, 2],
        [2, 0, 0, 0, 1, 3, 1, 1, 2, 3],
    ]
)

end = Board(
    [
        [3, 1, 2, 2, 3, 3, 3, 0, 1, 3],
        [0, 1, 0, 2, 2, 1, 3, 1, 0, 2],
        [1, 0, 1, 2, 0, 3, 0, 3, 0, 0],
        [3, 3, 3, 1, 1, 3, 3, 3, 3, 1],
        [2, 3, 0, 2, 2, 1, 2, 2, 3, 1],
        [1, 0, 2, 0, 0, 0, 1, 0, 0, 0],
        [1, 0, 3, 0, 3, 3, 1, 0, 2, 0],
        [0, 2, 1, 1, 0, 3, 2, 2, 2, 2],
        [2, 2, 2, 3, 0, 3, 1, 2, 2, 3],
        [0, 2, 3, 0, 0, 1, 2, 3, 3, 2],
    ]
)

_, fillone_actions = start.fillone(end)
actions = compress_actions(10, 10, fillone_actions)
actions
```

```python
# actionsをjsonにする
# "0" => "top",
# "1" => "bottom",
# "2" => "right",
# "3" => "left",

import json
json_dict = {}
json_dict["n"] = len(actions)
json_dict["actions"] = []
for i, action in enumerate(actions):
    if action[3] == "rowup":
        d = "0"
    elif action[3] == "down":
        d = "1"
    elif action[3] == "right":
        d = "2"
    elif action[3] == "left":
        d = "3"


    json_dict["actions"].append(
        {
            "p": action[2],
            "x": action[0],
            "y": action[1],
            "s": d
        }
    )

with open("fillone.json", "w") as f:
    json.dump(json_dict, f, indent=4)
```
